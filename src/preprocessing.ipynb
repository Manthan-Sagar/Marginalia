```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'../processed_book_data_kaggle.csv')
df.head()
```

```python
df = df.drop(columns=["tags"])
```

```python
def clean_category_names(df):
    if 'category_names' in df.columns:
        # Replace " | " with " "
        df['category_names'] = df['category_names'].str.replace('|', ' ', regex=False)
    return df

# Usage
df = clean_category_names(df)
df.head()
```

```python
def lowercase_columns(df):
    cols_to_lower = ['author_names', 'category_names', 'description']
    
    for col in cols_to_lower:
        if col in df.columns:
            # maintain NaN values as they are, only lowercase strings
            df[col] = df[col].str.lower()
            
    return df

# Usage
df = lowercase_columns(df)
df.head()
```

```python
import re

def clean_description_text(df):
    def clean(text):
        if not isinstance(text, str):
            return ""
        # Remove specific unwanted characters using Regex
        # This pattern keeps: letters (a-z), numbers (0-9), and spaces (\s)
        return re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Apply the cleaning function to the description column
    df['description'] = df['description'].apply(clean)
    return df

# Usage
df = clean_description_text(df)
df.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Ensure necessary NLTK data is downloaded
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt') # For tokenization if using word_tokenize, though split() is often enough for simple cases.

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    if not isinstance(text, str):
        return ""
    # Split text into words (tokens)
    words = text.split()
    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    # Join back into a string
    return " ".join(lemmatized_words)

# Apply to description column
df['description'] = df['description'].apply(lemmatize_text)

# Check the results
df[['description']].head()
```

```python
df.iloc[5].description
```

```python
# Create new column "tags"
# We use fillna('') to handle missing values safely
df['tags'] = (
    df['description'].fillna('') + ' ' + 
    df['author_names'].fillna('') + ' ' + 
    df['category_names'].fillna('') + ' ' + 
    df['title'].fillna('')
)

# Optional: Clean up extra spaces resulting from empty columns
df['tags'] = df['tags'].apply(lambda x: " ".join(x.split()))

# Check the result
df['tags'].head()
```

```python
df.iloc[0].tags
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Configuration
tfidf = TfidfVectorizer(
    stop_words='english',
    ngram_range=(1, 2),    # Captures "unrequited love" as a phrase
    max_features=20000,
    min_df=2,
    max_df=0.85
)

# Fit and transform the 'tags' column
tfidf_matrix = tfidf.fit_transform(df['tags'])

print(f"Matrix shape: {tfidf_matrix.shape}")
```

```python
import pickle
import scipy.sparse
import json

# 1. Save the trained TF-IDF Vectorizer
with open('vectorizer.pkl', 'wb') as f:
    pickle.dump(tfidf, f) # 'tfidf' is your fitted TfidfVectorizer object

# 2. Save the Document-Term Matrix (Sparse Matrix)
scipy.sparse.save_npz('book_vectors.npz', tfidf_matrix)

# 3. Save the ID mapping (Title -> Matrix Row Index)
# This allows you to look up which row belongs to which book later
# We use 'title' as the ID for now, but you could use 'isbn13' if preferred
id_to_index = {title: idx for idx, title in enumerate(df['title'])}

with open('id_to_index.json', 'w') as f:
    json.dump(id_to_index, f)

print("Artifacts saved: vectorizer.pkl, book_vectors.npz, id_to_index.json")

# To load:
# tfidf = pickle.load(open('vectorizer.pkl', 'rb'))
# tfidf_matrix = scipy.sparse.load_npz('book_vectors.npz')
# id_map = json.load(open('id_to_index.json', 'r'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity


def search_books(query, vectorizer, book_vectors, df, top_n=5):
    # 1. Vectorize query
    query_vector = vectorizer.transform([query])

    # 2. Cosine similarity
    scores = cosine_similarity(query_vector, book_vectors).flatten()

    # 3. Top N indices
    top_indices = scores.argsort()[-top_n:][::-1]

    # 4. Results
    results = df.iloc[top_indices].copy()
    results["similarity_score"] = scores[top_indices]

    return results

```

```python
results = search_books(
    "wizard magic school",
    tfidf,
    tfidf_matrix,
    df,
    top_n=5
)

print(results[["title", "similarity_score"]])
```
