```python
author_map = pd.Series(authors_df.author_name.values, index = authors_df.author_id).to_dict()
```

```python
import ast
def get_author_names(author_ids_str):
    try:
        # Convert string "[1, 2]" to actual list [1, 2]
        author_ids = ast.literal_eval(author_ids_str)
        
        # specific check if it's not a list (e.g. NaN or other formats)
        if not isinstance(author_ids, list):
            return ""
            
        names = []
        for aid in author_ids:
            # Get name from map, default to "Unknown" if not found
            name = author_map.get(aid, "Unknown")
            if isinstance(name, str):
                names.append(name)
        
        return " | ".join(names) # Join multiple authors with a separator
    except:
        return ""
```

```python
df['author_names'] = df['authors'].apply(get_author_names)
```

```python
categories_df = pd.read_csv(r"data/books_dataset/categories.csv")
category_map = pd.Series(categories_df.category_name.values, index=categories_df.category_id).to_dict()
formats_df = pd.read_csv(r"data/books_dataset/formats.csv")
format_map = pd.Series(formats_df.format_name.values, index=formats_df.format_id).to_dict()
```

```python
def get_names_from_list(ids_str, mapping_dict):
    try:
        if pd.isna(ids_str): return ""
        ids = ast.literal_eval(ids_str)
        if not isinstance(ids, list):
            return ""
        
        names = []
        for i in ids:
            name = mapping_dict.get(i, "Unknown")
            if isinstance(name, str):
                names.append(name)
        return " | ".join(names)
    except:
        return ""

def get_name_from_id(id_val, mapping_dict):
    try:
        return mapping_dict.get(id_val, "Unknown")
    except:
        return ""
```

```python
df['category_names'] = df['categories'].apply(lambda x: get_names_from_list(x, category_map))
```

```python
df.head()
```

```python
df["format_name"]=df["format"].apply(lambda x: get_name_from_id(x,format_map))
```

```python
def clean_authors(text):
    if pd.isna(text):
        return ""
    # 1. Split identifying multiple authors by the pipe
    authors = text.split('|')
    
    # 2. For each author, remove the spaces (merge first & last name)
    # .strip() removes accidental surrounding whitespace first
    merged_authors = [a.strip().replace(" ", "") for a in authors]
    
    # 3. Join them back together with a single space
    return " ".join(merged_authors)

# Apply the function
df['author_names'] = df['author_names'].apply(clean_authors)

# Verify the result
print(df['author_names'].head())
```

```python
import re

# 1. Define a text cleaning function
def clean_and_combine(row):
    # Collect parts, treating NaNs as empty strings
    author = str(row['author_names']) if pd.notna(row['author_names']) else ''
    category = str(row['category_names']) if 'category_names' in row and pd.notna(row['category_names']) else ''
    description = str(row['description']) if pd.notna(row['description']) else ''
    
    # Concatenate them with spaces
    combined_text = f"{author} {category} {description}"
    
    # Lowercase everything
    combined_text = combined_text.lower()
    
    # Replace pipes '|' with space
    combined_text = combined_text.replace('|', ' ')
    
    # Remove specific unwanted characters using Regex
    # This pattern keeps: letters (a-z), numbers (0-9), and spaces (\s)
    # It removes: [] brackets, commas, quotes, special symbols like @#$% etc.
    combined_text = re.sub(r'[^a-zA-Z0-9\s]', '', combined_text)
    
    # Remove extra spaces (e.g. "   " becomes " ")
    combined_text = re.sub(r'\s+', ' ', combined_text).strip()
    
    return combined_text

# 2. Apply the function to create the new column
# (Make sure 'category_names' exists. If you named it 'categories', change it below!)
df['tags'] = df.apply(clean_and_combine, axis=1)

```
